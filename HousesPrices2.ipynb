```python
from copy import deepcopy

train_df_sub = deepcopy(train_df[['OverallQual','OverallQual','GarageCars','KitchenAbvGr','EnclosedPorch','MSSubClass','SalePrice']])
```

```python
train_df_sub
```

```python
train_df_sub = train_df_sub.fillna(train_df_sub.mean())
```

```python
train_df_sub.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(train_df_sub.drop('SalePrice', axis=1), train_df_sub['SalePrice'], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
from xgboost import XGBRegressor # Changed to XGBRegressor
from sklearn.tree import DecisionTreeRegressor # Changed to DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor # Changed to RandomForestRegressor
from sklearn.linear_model import LinearRegression # Kept as LinearRegression
from lightgbm import LGBMRegressor # Changed to LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error # Changed to mean_squared_error


# ... (rest of your code)

# Define the models
model1 = XGBRegressor(random_state=42) # Changed to XGBRegressor
model2 = DecisionTreeRegressor(random_state=42) # Changed to DecisionTreeRegressor
model3 = RandomForestRegressor(random_state=42) # Changed to RandomForestRegressor
model4 = LinearRegression() # Kept as LinearRegression
model5 = LGBMRegressor(random_state=42) # Changed to LGBMRegressor

# Train and evaluate each model
models = [model1, model2, model3, model4, model5]
model_names = ['XGBRegressor', 'DecisionTreeRegressor', 'RandomForestRegressor', 'LinearRegression', 'LGBMRegressor'] # Changed model names
mse_scores = [] # Changed to mse_scores

for model, name in zip(models, model_names):
    # Convert X_train to NumPy array before fitting
    model.fit(X_train.to_numpy(), y_train)
    y_pred = model.predict(X_test.to_numpy()) # Convert X_test to NumPy array before predicting
    mse = mean_squared_error(y_test, y_pred) # Changed to mean_squared_error
    mse_scores.append(mse)
    print(f'{name}: MSE = {mse}')

# Find the best model
best_model_index = mse_scores.index(min(mse_scores)) # Changed to find minimum MSE
best_model_name = model_names[best_model_index]
best_model_mse = mse_scores[best_model_index]

print(f'\nBest Model: {best_model_name} with MSE = {best_model_mse}')
```
